```python
# ploting the class distribution
plt.figure(figsize=(8,6))
sns.countplot(x='Class',data=credit,color='lime')
plt.title('class distrbution')
plt.xlabel('class')
plt.ylabel('count')
plt.xticks([0,1],['legitimate','fraud'])
plt.show()
```

Fraud detection will require special metrics like precision/recall not accuracy because it is imbalanced
 

```python
# pie charts
plt.figure(figsize=(5,5))
plt.pie(credit['Class'].value_counts(),labels=['legitimate','fraud'],autopct='%1.2f%%',colors=['#3498db', "#AE3535"],startangle=90, wedgeprops={'width': 0.7})
plt.title('Class distribution percentage')
plt.text(0, 0, 'Class', ha='center', va='center', fontsize=14, weight='bold')
plt.axis('equal')
plt.show()
```

```python
# amount distributions
plt.figure(figsize=(10,6))
sns.histplot(credit['Amount'],bins=50,kde=True,color='brown')
plt.title('amount distribution')
plt.xlabel('amount')
plt.ylabel('count')
plt.xscale('log')
plt.show()
```

Only few are very large transaction

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Class',y='Amount',data=credit)
plt.title('transaction amount and class')
plt.ylim(0,3000)
plt.show()
```

Fraud tends to occur in low value transactions

```python
# Fraust Percentage by Amount Buckets:
credit['amount_bin'] = pd.qcut(credit['Amount'], 10)

fraud_rate = credit.groupby('amount_bin')['Class'].mean()

fraud_rate.plot(kind='bar', figsize=(10,4))
plt.title("Fraud Rate by Amount Range")
plt.ylabel("Fraud Rate")
plt.show()
```

Fraud is not uniformly distributed across amounts

```python
# time distribution
plt.figure(figsize=(10,6))
sns.histplot(credit['Time'],bins=100,kde=True,color='lime')
plt.title('time distribution')
plt.xlabel('time (seconds)')
plt.show()

credit['Hour']=credit['Time']/3600 % 24

plt.figure(figsize=(12,7))
sns.histplot(data=credit,x='Hour',hue='Class',multiple='stack',bins=24,kde=True)
plt.title('transaction by hour of day ( fraud vs legitimate)')
plt.xlabel('hour of the day')
plt.show()
```

Fraud not strongly time-dependent but some patters exists 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Select only V features
important_features = [col for col in credit.columns if col.startswith('V')]

n_features = len(important_features)
n_cols = 3
n_rows = math.ceil(n_features / n_cols)

plt.figure(figsize=(15, n_rows * 3))

for i, feature in enumerate(important_features):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.histplot(credit[feature], bins=50, kde=True)
    plt.title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

features_list = ['V15','V13','V26','V22','V23','V24',
                 'V28','V20','V27','V21','V8','V19']

plt.figure(figsize=(18, 12))

for i, feature in enumerate(features_list):
    plt.subplot(4, 3, i + 1)   # 4x3 grid = 12 plots
    
    sns.kdeplot(
        credit.loc[credit['Class'] == 0, feature],
        label='Legitimate',
        fill=True
    )
    sns.kdeplot(
        credit.loc[credit['Class'] == 1, feature],
        label='Fraud',
        fill=True
    )
    
    plt.title(f'{feature}: Fraud vs Legitimate')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# correlation matrix 
plt.figure(figsize=(25,20))
correlation_matrix = credit.corr()
sns.heatmap(correlation_matrix,cmap='coolwarm',annot=True,linewidths=1)
plt.title('correlation matrix of all features')
plt.tight_layout()
plt.show()
```

The correlation shows a strong diagonal,indicating perfect self-correlation for each feature, which is expected. Most off_diagonal values are close to zero, suggesting very low correlation betwee different features.  This indicates that the features are largely independent ,likely due to pca transformation.  There is no significat multicollinearity present.
